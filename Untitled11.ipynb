```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load dataset
df = pd.read_csv("/content/healthcare_dataset (1).csv")
```

```python
# Drop 'Loan_ID' column
df.drop(columns=['Name'], inplace=True)
```

```python
df
```

```python
# Fill missing values
# Fill categorical columns with mode
categorical_cols = ['Gender', 'Blood Type', 'Medical Condition', 'Date of Admission', 'Doctor', 'Hospital', 'Insurance Provider', 'Room Number', 'Admission Type', 'Discharge Date', 'Medication', 'Test Results']
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
print(df.columns)
numerical_cols = ['Age', 'Billing Amount', 'Room Number', 'Length of Stay'] # Example: Changed to 'Length of Stay' based on a common pattern
for col in numerical_cols:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        print(f"Warning: Column '{col}' not found in DataFrame.")
```

```python
df.info()
```

```python
# Convert categorical values to numeric using .replace()
df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 0})
df['Blood Type'] = df['Blood Type'].replace({'A+':0, 'A-':1, 'B+':2, 'B-':3, 'AB+':4, 'AB-':5, 'O+':6, 'O-':7})
df['Medical Condition'] = df['Medical Condition'].replace({'Cancer':0, 'Obesity':1, 'Diabetes':2})
df['Admission Type'] = df['Admission Type'].replace({'Urgent':0, 'Emergency':1, 'Elective':2})
df['Test Results'] = df['Test Results'].replace({'Normal':0, 'Abnormal':1, 'Inconclusive':2})
display(df.head())
```

```python
df['Arthritis'] = pd.to_numeric(df['Arthritis'], errors='coerce')
```

```python
# Split into features and target
X = df.drop('Test Results', axis=1)
y = df['Test Results']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```
